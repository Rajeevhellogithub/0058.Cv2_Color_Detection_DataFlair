```python
import cv2
import numpy as np

def arrange_frames(frames, grid_size):
    # Calculate the size of each frame
    frame_height, frame_width = frames[0].shape[:2]
    
    # Create a blank canvas for the grid
    grid_height = frame_height * grid_size[0]
    grid_width = frame_width * grid_size[1]
    grid = np.zeros((grid_height, grid_width, 3), dtype=np.uint8)

    # Place each frame in the grid
    for idx, frame in enumerate(frames):
        row = idx // grid_size[1]
        col = idx % grid_size[1]
        y_start = row * frame_height
        y_end = y_start + frame_height
        x_start = col * frame_width
        x_end = x_start + frame_width
        
        # Resize frame if necessary
        if frame.shape[0] != frame_height or frame.shape[1] != frame_width:
            frame = cv2.resize(frame, (frame_width, frame_height))
        
        grid[y_start:y_end, x_start:x_end] = frame

    return grid

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Red color
    low_red = np.array([161, 155, 84])
    high_red = np.array([179, 255, 255])
    red_mask = cv2.inRange(hsv_frame, low_red, high_red)
    red = cv2.bitwise_and(frame, frame, mask=red_mask)

    # Blue color
    low_blue = np.array([94, 80, 2])
    high_blue = np.array([126, 255, 255])
    blue_mask = cv2.inRange(hsv_frame, low_blue, high_blue)
    blue = cv2.bitwise_and(frame, frame, mask=blue_mask)

    # Green color
    low_green = np.array([25, 52, 72])
    high_green = np.array([102, 255, 255])
    green_mask = cv2.inRange(hsv_frame, low_green, high_green)
    green = cv2.bitwise_and(frame, frame, mask=green_mask)

    # Every color except white
    low = np.array([0, 42, 0])
    high = np.array([179, 255, 255])
    mask = cv2.inRange(hsv_frame, low, high)
    result = cv2.bitwise_and(frame, frame, mask=mask)

    # Arrange frames in a grid
    frames = [frame, red, blue, green, result]
    grid_image = arrange_frames(frames, (2, 3))  # 2 rows and 3 columns

    # Display the grid of frames
    cv2.imshow("Frames in Grid", grid_image)

    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
